## Distribución Uniforme

Para generar números aleatorios uniformes entre dos valores dados usando scipy, podemos usar el siguiente código o el código que hemos
discutido en clase:

```python
def generate_uniform_random_numbers(bottom, ceiling, count):
    """
    Genera una lista de números aleatorios distribuidos uniformemente.
    
    Parámetros:
    bottom (float): el límite inferior de los números generados
    ceiling (float): el límite superior de los números generados
    count (int): cuántos números generar
    
    Devuelve:
    list: una lista de números aleatorios distribuidos uniformemente
    """
    return np.random.uniform(bottom, ceiling, count)

# Parámetros de la distribución uniforme
a = 0  # Límite inferior (inclusivo)
b = 10 # Límite superior (inclusivo)
size = 100 # Cantidad de números aleatorios a generar

# Generar números aleatorios
numeros_aleatorios = generate_uniform_random_numbers(a, b, size)

plt.hist(numeros_aleatorios, bins=20
         , density=True)
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribución Uniforme (0, 10)')
plt.show()
```

```python
from scipy.stats import norm 
```

```python
def generate_normal_random_numbers(mean, std_dev, count):
    """
    Genera una lista de números aleatorios distribuidos normalmente.
    
    Parámetros:
    mean (float): la media de la distribución
    std_dev (float): la desviación estándar de la distribución
    count (int): cuántos números generar
    
    Devuelve:
    list: una lista de números aleatorios distribuidos normalmente
    """
    return np.random.normal(mean, std_dev, count)
```

```python
# Generar números aleatorios para los conjuntos de parámetros dados
set1 = generate_normal_random_numbers(10, 1, 1000)
set2 = generate_normal_random_numbers(10, 50, 1000)
```

```python
# Graficar los histogramas
plt.figure(figsize=(12, 6))

# Histograma para el primer conjunto
plt.subplot(1, 2, 1)
plt.hist(set1, bins=30, edgecolor='black')
plt.title('Distribución Normal: media=10, desviación típica=1 (1000 números)')
plt.xlabel('Valor')
plt.ylabel('Recuento')

# Histograma para el segundo conjunto
plt.subplot(1, 2, 2)
plt.hist(set2, bins=30, edgecolor='black')
plt.title('Distribución Normal: media=10, desviación típica=50 (1000 números)')
plt.xlabel('Valor')
plt.ylabel('Recuento')

plt.tight_layout()
plt.show()
```

## Distribución Exponencial

1. Utilizando `numpy.random.exponential`, crea una función que devuelva una lista de números distribuidos exponencialmente con la media de 10. 

1. Utiliza la función para generar dos secuencias de números de tamaño 10 y 100.

1. 2. Traza las distribuciones como histogramas con el número de bins como 100.

El resultado debe ser como el siguiente:



```python
import numpy as np
import matplotlib.pyplot as plt

def generate_exponential_random_numbers(mean, size):
    """
    Genera una lista de números aleatorios distribuidos exponencialmente.
    
    Parámetros:
    mean (float): la media de la distribución exponencial
    size (int): tamaño de la muestra
    
    Devuelve:
    list: una lista de números aleatorios distribuidos exponencialmente
    """
    return np.random.exponential(scale=mean, size=size)

# Generar números exponenciales para dos secuencias diferentes
sequence1 = generate_exponential_random_numbers(10, 10)
sequence2 = generate_exponential_random_numbers(10, 100)

# Graficar los histogramas
plt.figure(figsize=(12, 6))

# Histograma para la primera secuencia
plt.subplot(1, 2, 1)
plt.hist(sequence1, bins=100, edgecolor='black')
plt.title('Distribución Exponencial: media=10, tamaño=10')
plt.xlabel('Valor')
plt.ylabel('Recuento')

# Histograma para la segunda secuencia
plt.subplot(1, 2, 2)
plt.hist(sequence2, bins=100, edgecolor='black')
plt.title('Distribución Exponencial: media=10, tamaño=100')
plt.xlabel('Valor')
plt.ylabel('Recuento')

plt.tight_layout()
plt.show()
```

¿En qué se diferencian ambas distribuciones?

La media cambia, por lo que la distribución también cambia.
